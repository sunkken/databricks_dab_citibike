```python
from pyspark import pipelines as dp
from pyspark.sql.functions import col, unix_timestamp, to_date
```

```python
@dp.table(
    comment="Silver layer: Cleaned and transformed Citibike trip data"
)
def silver_citibike():
    return (
        spark.read.table("bronze_citibike") \
        .withColumn("trip_duration_mins", (unix_timestamp(col("ended_at")) - unix_timestamp(col("started_at"))) / 60) \
        .withColumn("trip_start_date", to_date(col("started_at"))) \
        .select(
            "ride_id",
            "trip_start_date",
            "started_at",
            "ended_at",
            "start_station_name",
            "end_station_name",
            "trip_duration_mins"
        )
    )
```
